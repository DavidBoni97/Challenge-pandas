### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda = round(tienda["Precio"].sum())
ingreso_total_tienda2 = round(tienda2["Precio"].sum())
ingreso_total_tienda3 = round(tienda3["Precio"].sum())
ingreso_total_tienda4 = round(tienda4["Precio"].sum())

print(f"El ingreso total de la tienda 1 es de {ingreso_total_tienda}")
print(f"El ingreso total de la tienda 2 es de {ingreso_total_tienda2}")
print(f"El ingreso total de la tienda 3 es de {ingreso_total_tienda3}")
print(f"El ingreso total de la tienda 4 es de {ingreso_total_tienda4}")
```

```python
tienda['Beneficio'] = tienda['Precio'] - tienda['Costo de envío']
tienda2['Beneficio'] = tienda2['Precio'] - tienda2['Costo de envío']
tienda3['Beneficio'] = tienda3['Precio'] - tienda3['Costo de envío']
tienda4['Beneficio'] = tienda4['Precio'] - tienda4['Costo de envío']
```

```python
beneficio_total_t1 = round(tienda['Beneficio'].sum())
beneficio_total_t2 = round(tienda2['Beneficio'].sum())
beneficio_total_t3 = round(tienda3['Beneficio'].sum())
beneficio_total_t4 = round(tienda4['Beneficio'].sum())

print(f'El beneficio total de la tienda 1 fue de: {beneficio_total_t1}')
print(f'El beneficio total de la tienda 2 fue de: {beneficio_total_t2}')
print(f'El beneficio total de la tienda 3 fue de: {beneficio_total_t3}')
print(f'El beneficio total de la tienda 4 fue de: {beneficio_total_t4}')
```

```python
beneficios_totales = {
                                    'Tienda 1': beneficio_total_t1,
                                    'Tienda 2': beneficio_total_t2,
                                    'Tienda 3':beneficio_total_t3,
                                    'Tienda 4': beneficio_total_t4
                                }

df_beneficios_totales = pd.DataFrame.from_dict(beneficios_totales, orient='index', columns=['Beneficio Total'])
df_beneficios_totales
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_beneficios_totales.index, df_beneficios_totales['Beneficio Total'], color = 'green')
plt.xlabel('Tiendas')
plt.ylabel('Beneficio Total')
plt.title('Comparación de Beneficio Total por Tienda')
```

#7. Conclusion

La tienda que debería venderse es la Tienda 4. Aunque sus calificaciones promedio y la distribución de artículos vendidos por categoría son similares al resto de las tiendas, presenta una desventaja significativa en su beneficio neto (precio − costo de envío). La Tienda 4 obtiene consistentemente los beneficios más bajos, lo que indica un menor rendimiento económico pese a tener un comportamiento similar en otros indicadores.

Dado que las demás tiendas muestran beneficios superiores bajo condiciones prácticamente equivalentes en ventas y satisfacción del cliente, la Tienda 4 se convierte en la opción menos eficiente desde el punto de vista financiero. Por lo tanto, la decisión más adecuada es vender la Tienda 4 para optimizar el rendimiento global de la cadena.
